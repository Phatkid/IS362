# College Masters Programs Unemployment Stats

## Imports and Options

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', 50)
pd.set_option('display.max_rows', 1000)
```

```python
URL = "https://raw.githubusercontent.com/fivethirtyeight/data/master/college-majors/all-ages.csv"
df = pd.read_csv(URL) # Reads csv from web
df.shape # shows shape of date rows and columns
```

```python
df.head() # checks data read
```

### Determine what data will be review based on the field in data

```python
df['Major_category'].unique() # shows the different categories in this column
```

```python
df1 = df[['Major_category', 'Major', 'Unemployment_rate']] # New dataframe with relevant columns
df1.head()
```

### Display the top five Unemployment rates by major

```python
highest_unem = df1.sort_values(['Unemployment_rate'], ascending=False).head(5) 
highest_unem
```

### Compare the data for Major Categories "Education" and "Social Sciences"

```python
df2 = df1[df1['Major_category'].isin(['Social Science', 'Education'])] # new dataframe for Educaction and Social Science
df2
```

### Compare mean for both of the chosen major categories

```python
grouped_mean = df2.groupby('Major_category').mean() # displays mean for major categories
grouped_mean
```

```python
grouped_med = df2.groupby('Major_category').median() # displays median for major category
grouped_med
```

### Compare Majors with Visualization 

```python
ax = df2.plot.bar(x= 'Major', figsize=(14,10), fontsize=14, title="Unemployment Rates", legend=False)
ax.set_ylabel('Unemployment Rate')
```

## Conclusion

** School Student Counseling and Library Sciences have the highest unemployment rate by major. As a major category, Social Science has a higher mean and median unemploymet rate. **
