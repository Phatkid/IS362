## Task 4

Some of the horsepower data is marked as missing with a ‘?’. You need to replace this ‘?’ with a pandas-appropriate missing value, then convert the column to numeric.

```python
(df.horsepower == '?').sum() # prior to replacing with np.nan
```

```python
df = df.replace('?', np.nan) # replaces '?' with np.nan
```

```python
(df.horsepower.isnull()).sum() # sum of value matches prior test
```

```python
(df.horsepower == '?').sum() # confirms the '?' have been replaced by showing a 0 sum
```

```python
df['horsepower'] = df.horsepower.apply(pd.to_numeric, errors='ignore')
```

```python
df.dtypes # confirms that the horsepower column is numeric
```

## Task 5

The origin column values are 1, 2, and 3. Convert these values respectively to ‘USA’, ‘Asia’, and ‘Europe’

```python
df.head(20) # prior to replace values in the origin column
```

```python
df['origin'].replace(1, 'USA', inplace=True)
df['origin'].replace(2, 'ASIA', inplace=True)
df['origin'].replace(3, 'EUROPE', inplace=True)
```

```python
df.head(20) # Checks for replace values in the origin column
```

## Task 6

Create a bar chart that shows the distribution for cylinders.

```python
df.groupby('cylinders').size().plot(x = 'cylinders', kind='bar', figsize = ( 12 , 10 ), title='CYLINDERS')
```

## Task 7

Create a scatterplot that shows the relationship between horsepower and weight.

```python
x1 = df['horsepower']
y1 = df['weight']
ax = sns.regplot(x= x1, y= y1, )
ax.set(xlabel='Horespower', ylabel='weight', title= 'Cars')
```
